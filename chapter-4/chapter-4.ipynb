## Loops: percorendo uma lista inteira

Por muitas vezes queremos oercorer um lista por diversos fatores como realuizar um funcao estatistica por todas os elementos ou simplesmente exibir todos os valores es destaque. Imagine que temos uma lista com nomes de magicos famosos e queremos exibir todos individualmente.

```python
magicians = ['alice', 'david', 'carolina']
```

No metodo que aprendemos ate agora presisariamos usar o printe em cada posicao de lista

```python
print(magicians[0])
print(magicians[1])
print(magicians[2])
```

e se a lista amudar de tamanho? e se tivermos mais de 100 nomes de magicos? nao e viavel usar a funcao `print` 100 vezes ou mudar o codigo toda vez que o tamnhai de lista mudasse.

Para isso temos a funcao `for` que percorre toda a nossa lista nao importe o quanto ela mude ou o tamanho dela. A estrutura e bem simples primeiro sempre vem o `for` depois algo que represente cada elemnto no nossa casa magico depois vem `in` depois a nossa lista magicos por fim os `:`. desse forma enviamos um comando para o python assim `para cada elemeto magico dentro da lista magicos exiba o nemo do magico`.

```python
for magician in magicians:
    print(magician)
```

Dessa forma o computador auto matisa algumas tarefas repetitivas.
dessa forma ele repete o comando `print` ate o ultimo valor da nossa lista

`IMPORTANTE`:
- 1 se nossa lista estiver com 1.000.000 de nomes ele exibirar 1.000.000 de nomes de magicos.
- 2 uma boa preatica nos `loops` e usar nomes representativos por exemplo **for** `gato` **in** `gatos`: por exemplo. Uma maneira facil de lembrar seria colocar os nomes das listas sempre no plurar e o valore representativo sempre no simgular.
- 3 `Cuidado` com a `identacao`. abordaremos mais sobre isso um pouco mais a frente

## Fazer mais tarefas dentro de um loop

Podemor fazer praticamnete qualquer cuisa dentro do nosso `for`.

usando a funcao para demonstrar `.title()`

```python
for magician in magicians:
    print(f"{magician.title()}, that was a great magician!")
```

Nesse casao suamos a funcao `.title()` para deixar a primeira letra maiuscula e mandar uma mensagem para cada magico

Podemos escrever quantas linhas de codigo quisermos identada (antes da linha vertical do `for`)

<pre>
for magician in magicians:
    ➜ command to be executed by the for loop
    ➜ another command

code not executed by the for loop
</pre>

```python
for magician in magicians:
    print(f"{magician.title()}, that was a great magician!")
    print(f"I can't wait to see your next trick {magician.title()}.\n")
```

Percebas que alendo do primeiro comando adicionamos o segundo. `print(f"I can't wait to see your next trick {magician.title()}.\n")` e o `\n` para exibir um espaco vazio entre os a exibicao dos `prints`. E possivel fazer quantos comandos quisermos dentro da identacao do `for` que for nescecario

## Fazer mais tarefas apos usar o loop do `for`

O que fazer apos terminar o loop `for`? apos terminar o loop passa para o procimo passo do programa. como exeplo vamos exibir uma mensagem de agradecimento para todos depois que o loop terminar

```python
for magician in magicians:
    print(f"{magician.title()}, that was a great magician!")
    print(f"I can't wait to see your next trick {magician.title()}.\n")

print("Thank you everyone, that was a great show!")
```

perceba que todos os printse de cada magico dentro da identacao foram executados. somente o ultimo que nao esta dentro da identacao foi mostrado. opara vermos a diferenca vamos identar o ultimo print para dentro da identacao do for

```python
for magician in magicians:
    print(f"{magician.title()}, that was a great magician!")
    print(f"I can't wait to see your next trick {magician.title()}.\n")

    print("Thank you everyone, that was a great show!")
```

veja que a ultimo printe foi executado multiplas vezes. Por isso temos de tomar cuidade para `nao executar multiplas tarefas desnecesaria`

### `range()`

A funcao `range()` pode facilitar nosso trabalho quando quisermos criar uma lista de numeroes. imagine que queremos gerar uma serie de numeros. em vez de criar uma lista manualmnte e aplicar o `for` podemos usar o `range()`.

```python
for value in range(1,5):
    print(value)
```

isso nos poupa tarbalho uma vez que se precisarmos gerar numeras de 1 a 1.000.000 por exemplo

Perceba que o nosso codigo nao gerou o numero 5 isso tartase do resultado de comprimento `off-by-one` dentre muita que voce possivelmente verar nas linguagens de programacao no `range` ele comeca a comtagem no primeiro valor e para no ultimo se quisermos ate o valor 5 teramos que colocar o `range`
ate o valor 6

```python
for value in range(1,6):
    print(value)
```

Voce pode usar o range em diferentes intervalos e comprimentos. mas por padar se voce usar colocar somnete um, valor dentro do `range` ele atribui o primeiro a 0.

```python
for value in range(6):
    print(value)
```

### Usando um range para criar uma lista numerica

E possivel criarmos uma lista numerica `convertendo` o `range` em um lista usando a funcao `list()` quando colocamos o `range()` dentro das chave e convertido para uma lista

```python
numbers = list(range(1,6))
print(numbers)
```

No range podemos colocar mais um argumento dpois da nossa parada se quisermos um determinado ientervalo podendo ficar assim `range(inicio, fim, intervalo)`

se quisermos somnete valore pares

```python

numbers = list(range(2,10,2))
print(numbers)
```

Ou valores multiplos de 10

```python
numbers = list(range(10,201,10))
print(numbers)
```

`IMPORTANTE`:
- 1 se voce atribuir somente um argumento no range ala porm padram define o valor informado e o segundo e o primerio ele assume que e o `0`
- 2 se voce atribui dois argumentos no range ele assume que o primeiro e o ponto de partida e o ultimo a parada
- 3 se colocarmos tres argumneto no range o terceiro argumneto fico com um intervalo dos valore a ser gerado

Podemos tambem criar uma lista de valore quadrados dom o comando no python `**` junto com o `for` e o `range`

```python
squares = []
for number in range(1, 11):
    square = number ** 2
    squares.append(square)

print(squares)
```

Note que primeiro criamos uma lista vazia, depois definimos o nosso for com o range com o intervalo desejado. doies usamo o `**` para velar o valor o numero ao quadrado e atribuinda a uma variavel temporaria quadrado e por fim armasemnao na lista vazia.
poderiams ter feito a operacao de elevar o numero ao quadrado somente drietemnte dentro da funcao `.appende`

```python
squares = []
for number in range(1, 11):
    squares.append(number ** 2)

print(squares)
```

## List comprehensions

Quando criamos a nossa lista de valores quadrados usamos 3 ou 4 linhas de comandos para gerar nossa lista. Com o `List comprehension` podemos usar somnete uma.

```python
squares = [value ** 2 for value in range(1,11)]
print(squares)
```

Para criar uma `list comprehension` basta primero digitar um nome descritivo para receber ser atribuido a nossa lista `=` dpeis usar dois colchetes `[]` e dentro dele primeiro vai a exprecao `suqare**2` a ser executada depois a estrutura normal do `for` que ja vimos antesmas `for value in range(1,11)` sem os `:`. E nessesario pratica para dominar a `List comprehensions` mas quando voce dominar a tecnica vera que seu codigo ficara bem mais limpo menor e facil de ler alen de facilitar seu trabalho.

## Trabalhando com partes de uma lista

No capitulo antirior trabalhamos com a lista toda mais o Python nos oferece a possibilidade de trabalhar com partes de uma lista que e chamado de `slicing`. que pode ser usado tamto com listas de valores numericos quanto strings. O `slicing` funciona muito parecido com o `range` do python.

```python
squares = [value ** 2 for value in range(1,31)]
print(squares)
```

Sabemos que se clocarmos `[um valor numerico]` logo apos a ums lista no especificamos o valor que queremos mas se clocarmo `:` antes desse valor...

```python
squares[:15]
```

Perceba que ele retornou todos so valores da primeira posicao 0 ate a posicao 2 muito parecido com a funcao `range`, se colocarmos outro argumneto antes dos dois pontos e mantiver o segundo argumneto teremos uma fatia da lista.

```python
squares[10:15]
```

nesse caso pegamos a 10 posicao ate a 15 posicao. e se tirarmos o segundo argumnto e mantiver o primeiro temos

```python
squares[10:]
```

O resta da lista apartir da 10 posicao. lembrece que se usarmos valore negativos a contagem comeca do ultimo elemeto da lista>


```python
squares[-3:]
```

Nesse caso pegamos somente as 3 ultimas poscoes da lista, um truque a mais do python e que podemos inverter a lista usando a simtax `[::-1]`

```python
squares[::-1]
```
