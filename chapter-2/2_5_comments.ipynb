## 2.5) Comments

Comments are an essential part of programming. They allow you to write notes in your code that are ignored by the Python interpreter. Good comments help you and others understand your code, especially as it becomes more complex.

### How to Write a Comment

In Python, a comment starts with a hash mark (`#`). Anything following the `#` on that line is considered a comment and will not be executed.

```python
# This is a comment. It explains what the code below does.
# Say hello to everyone.
print("Hello Python people!")
```

### When to Write Comments

The main reason to write comments is to explain **why** you are doing something, not just **what** you are doing. If your code is complex or has a non-obvious purpose, a comment can save a lot of time for anyone reading it later (including your future self!).

If a piece of code is self-explanatory (like `print("Hello World!")`), it doesn't need a comment. However, if the logic is tricky, a short explanation can be invaluable.

Writing good comments is a skill that you will develop over time. For now, the most important thing is to use them to clarify parts of your code that you think might be hard to understand when you come back to them.
