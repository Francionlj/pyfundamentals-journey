In this chapter, we will learn how to write **functions**, which are named blocks of code designed to perform a specific task. Instead of rewriting the same code multiple times, you can define a function once and call it whenever you need it. This makes your programs easier to write, read, test, and fix.

## 8.2) Passando argumentos

Uma funcao pode ter doversos parametros os mais doversos possiveis e pode precisar de muitos ardimnetos para funcionar. Podemos passar argumentos de uma funcao de diversas formas como:
- posicionais que pode passar o argumento exatamnete na ordem dos parametros
- argumentos nomeados em que cada argumento e composto de um nome de variavel.

### 8.2.1) argumentos posicionais

Ao chamarmos uma funcoa o python verifica a correspondencia de cada argumento na chamada da funcao. essa e a forma mais simples de se faser etomar comobase a ordem dos argumentos conhecidos como por exemplo

```python
def describe_pet(animal_type, pet_name):
    """"exibe a informacao sobre o animal de estimacao"""
    print(f"\nI have a {animal_type}.")
    print(f"\nMy {animal_type}'s is {pet_name}.")
```

```python
describe_pet("hamster", "harry")
```

#### 8.2.1.1) Multipas chmadas de funcoes

essa funcao revela que para funcionar precisamos de dois argumnetos o animal e o nome do animal. quando chamamos `describe_pet()`, precisamos nessesariamnete chemar usar os argumnetos na mesma ordem dos parametros. dodemos chamar essa funcao quantas vezes precisarmos por exemplo


```python
def describe_pet(animal_type, pet_name):
    """"exibe a informacao sobre o animal de estimacao"""
    print(f"\nI have a {animal_type}.")
    print(f"\nMy {animal_type}'s is {pet_name.title()}.")
```

```python
describe_pet("hamster", "harry")
describe_pet("dog", "willie")
```

A orem e extremamete importante para argumentos posicionais podemdo ter resultaos inesperados por exemplo se torcarmos a ordem
