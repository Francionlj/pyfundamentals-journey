Most programs are written to solve an end user's problem. To do so, you usually need to get some information from the user. In this chapter, we will learn how to accept user input using the `input()` function and how to keep programs running as long as necessary using `while` loops.

## 7.1) Apresentando os loops while

O loop `for` itera em uma colecao de itens e executa um bloco de codigo por vez ate que a colecao acabe. O `while` executa ate uma condicao aconteca por exempo ate uma contagem de um numero seja ate nemor ou igual a 5

```python
current_number = 1
while current_number <= 5:
    print(current_number)
    current_number += 1
```
