## 6.2) Trabalhando com dicionarios

O dicionario no Python e uma associacao de `chaves-valores` a onde topas as chaves te um valor especificp. O valor dessa chave pode ser qualquer obejeto no python como strings, numeros, uma lista ou ate mesmo outro dicionario.

No Python o dicionairo e definida por cchaves `{}` com um pares de `chaves-valores` `{valor:chave}`

```python
alien_0 = {'color': 'green', 'points': 5}
```

O conjunto de chaves e valore estao associadops entre si ao fornecer uma chave o python retorna o valor acossiado a essa chave. Cada chave  eassociadoa um valor por `:` e os pares de valor de `chave-valor` e separada por virgula `{cahe_1:valor_1 , chave_2:valor_2}`

### 6.2.1) Acessabdo valore em um dicionario

Para acessar valores de um didionario barta informar o nome do dicionario, e entre colchetes `[]` a chave que deseja:

```python
alien_0 = {'color': 'green', 'points': 5}
```

```python
print(alien_0['color'])
```

Nesse caso o python retornou o valor acosiado a `color` do dicionaio alien_0.

Um dicionario pode acessar um valor ilimitado de `chaves`. agora podemos se caso um jogador abater esse alien em especifico podemos exibira pontuacao que foi obtido.

```python
new_points = alien_0['points']
print(f"You just earned {new_points} points!")
```

Uma fez definido podemos extrair a valor da chave em especifico e utilisala ou exibila.
