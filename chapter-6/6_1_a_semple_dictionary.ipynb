# Chapter 6: Dictionaries

This notebook introduces **dictionaries**, a powerful Python data structure that allows you to connect related pieces of information. You will learn how to access, add, and modify data within a dictionary. Dictionaries can store an almost limitless amount of information, enabling you to model real-world objects accurately.

## 6.1) A Simple Dictionary

Consider a simple game featuring aliens that can have different colors and point values. A dictionary is the perfect way to store this information for a single alien.

```python
alien_0 = {'color': 'green', 'points': 5}

print(alien_0['color'])
print(alien_0['points'])
```

In this example, the dictionary `alien_0` stores the alien's color and point value. The `print()` statements access and display the value associated with each key.

Understanding dictionaries is essential for modeling complex data in Python, as they allow you to represent objects with multiple attributes in a structured way.
