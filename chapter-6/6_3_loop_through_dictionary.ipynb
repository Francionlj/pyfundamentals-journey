# Chapter 6: Dictionaries

This notebook demonstrates efficient ways to **loop through dictionaries** to access keys, values, or both.

## 6.3) Looping Through a Dictionary

### 6.3.1) Looping Through All Key-Value Pairs

Use the `.items()` method to access both keys and values.

```python
user_0 = {
    'username': 'efermi',
    'first': 'enrico',
    'last': 'fermi',
    }

for key, value in user_0.items():
    print(f"\nKey: {key}")
    print(f"Value: {value}")
```

### 6.3.2) Looping Through All Keys

Use the `.keys()` method to access only the keys. This is also the default behavior if you loop through the dictionary directly.

```python
favorite_languages = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'rust',
    'phil': 'python',
    }

for name in favorite_languages.keys():
    print(name.title())
```

### 6.3.3) Looping Through Keys in Order

To loop through keys in a specific order, use the `sorted()` function.

```python
for name in sorted(favorite_languages.keys()):
    print(f"{name.title()}, thank you for taking the poll.")
```

### 6.3.4) Looping Through All Values

Use the `.values()` method to access only the values.

```python
print("The following languages have been mentioned:")
for language in favorite_languages.values():
    print(language.title())
```

To avoid duplicates, wrap the values in `set()`. A **set** is a collection of unique items.

```python
print("The following languages have been mentioned:")
for language in set(favorite_languages.values()):
    print(language.title())
```
