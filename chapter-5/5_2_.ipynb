## 5.2) Testes condicionais

O python quando se usa o `if` ele avalia se o valor e `True`ou `False` para que ele possa decedir se a instrucao da condicional `if` sera ou nao executada.
* se o teste o valor for `True` o codigo sera executado
* se o teste o valor for `False` o codigo nao sera executado

### 5.2.1) Verificando igualdade

Boa parte dos teste do python compara o valor atual de uma variavel com um valor especifico de interece.

```python
car = 'bmw'
car == 'bmw'
```

Perceba que atribuir um valor a uma variavel e somete um simbolo de igualdade `=` e para comparar e dois simbolos de igualdade `==`. Tambe repare que a voler que o payton exibiu foi somnet `True` ou seja nesse exeplo `car = 'bmw'`. agora se car receber outra valor?

```python
car = 'audi'
car == 'bmw'
```
