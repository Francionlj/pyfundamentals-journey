# Chapter 5: If Statements

This notebook introduces **conditional statements** (`if`, `else`), which allow your program to respond to different situations. We will start with a simple example to understand how to perform an action based on a specific condition.

## 5.1) A Simple Example

The following example demonstrates a `for` loop that iterates through a list of cars. We want to print the name of one specific car, 'bmw', in all uppercase, while printing all other car names in title case.

```python
cars = ['audi', 'subaru', 'bmw', 'toyota']
for car in cars:
    if car == 'bmw':
        print(car.upper())
    else:
        print(car.title())
```

In this example, the loop checks each car's name. The `if` statement tests whether the current value of `car` is equal to `'bmw'`.

- If the test is `True`, the code inside the `if` block is executed, printing the name in uppercase.
- If the test is `False`, the code inside the `else` block is executed, printing the name in title case.

This simple `if-else` structure is the foundation for building complex logic in your programs and will be explored in more detail throughout this chapter.
